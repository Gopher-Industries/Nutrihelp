```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## FRUITS

```python
# Define the URL
url = "https://www.nutritiontable.com/nutritions/?c=11"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    
    div_element_units = soup.find("div", {"class": "vwtContainer widthContainer", "id": "pnlVwtContainer"})

    if div_element_units:
        # Find all <a> tags within this div
        a_tags = div_element_units.find_all("a")

        # Extract href attributes and corresponding text from the <a> tags
        href_lists = []
        
        for a in a_tags:
            href_list = a.get("href")
            # Append as a tuple
            href_lists.append(href_list)
```

```python
url_list = 'https://www.nutritiontable.com/nutritions/nutrient/?id=69'


# Send a GET request to the URL
response = requests.get(url_list)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
        
    fieldset_div = soup.find("div", {"id": "ctl00_cphMain_pnlFieldsetVW"})

    if fieldset_div:
        # Find all <a> tags within this div
        rowitem_divs = fieldset_div.find_all("div", {"class": "rowitem"})

        # Extract href attributes and corresponding text from the <a> tags
        row_header = ["Name",]
        for row in  rowitem_divs:
            text_name = row.get_text(separator='|', strip=True).split('|')[0]
            row_header.append(text_name)
```

```python
df = pd.DataFrame(columns=row_header)
```

```python
url_list = 'https://www.nutritiontable.com'

url = [f'{url_list}{link}' for link in href_lists]

for link in url:
    # Send a GET request to the URL
    response = requests.get(link)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, "html.parser")
        
        field_div = soup.find("div", {"id": "ctl00_cphMain_pnlShowProd", "class": "col-7 col-m-7 col-t-15"})
        if field_div:
            # Find all <a> tags within this div
            h2_tag = field_div.find("h2")
            fruit = h2_tag.text
        
        
        fieldset_div = soup.find("div", {"id": "ctl00_cphMain_pnlFieldsetVW"})

        if fieldset_div:
            # Find all <a> tags within this div
            rowitem_divs = fieldset_div.find_all("div", {"class": "rowitem"})

            # Extract href attributes and corresponding text from the <a> tags
            row = []
            row_value = []
            for row in  rowitem_divs:
                text_value = row.get_text(separator='|', strip=True).split('|')[1]
                row_value.append(text_value)
            
            row = [fruit] + row_value
            
            length = len(df)
            df.loc[length] = row
```

```python
# Define the URL
url = "https://www.nutritiontable.com/nutritions/?c=10"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find the <div> with class "mobspano" and id "ctl00_pnlHeaderLetters"
    div_element = soup.find("div", {"class": "headerLetters", "id": "ctl00_pnlHeaderLetters"})
    
    if div_element:
        # Find all <a> tags within this div
        a_tags = div_element.find_all("a")
        
        # Extract href attributes from the <a> tags
        href_list = [a.get("href") for a in a_tags]
```

```python
div_element_units = soup.find("div", {"class": "vwtContainer widthContainer", "id": "pnlVwtContainer"})

if div_element_units:
    # Find all <a> tags within this div
    a_tags = div_element_units.find_all("a")

    # Extract href attributes and corresponding text from the <a> tags
    lists = []
    for a in a_tags:
        href_list = a.get("href")
        href_text = a.text
        lists.append((href_list, href_text))  # Append as a tuple
```
