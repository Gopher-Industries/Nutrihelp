```python
# Importing necessary libraries
import pandas as pd
```

```python
# Read the CSV file into a DataFrame
fruits = pd.read_csv('friuts_nutritional_facts.csv')
fruits.head()
```

```python
# Display information about the DataFrame, including the data types of each column and memory usage
fruits.info()
```

```python
# Checking duplicate values
fruits.duplicated().sum()
```

```python
# Replace "-" with NaN
fruits.replace("-", float("nan"), inplace=True)
```

```python
# Replace commas with periods
columns_to_convert = fruits.columns.difference(['Name', 'Product group:'])
fruits[columns_to_convert] = fruits[columns_to_convert].replace(',', '.', regex=True)
```

```python
# Convert all columns except 'Name' and 'Product group' to float
fruits[columns_to_convert] = fruits[columns_to_convert].astype(float)
```

```python
# Rename columns to remove trailing colon
fruits.rename(columns=lambda x: x.split(':')[0].strip(), inplace=True)
```

```python
# Final in the required format
fruits.head()
```

```python
# Write the DataFrame to a CSV file
fruits.to_csv('fruits_nutritional_facts_preprocessed.csv', index=False)
```
